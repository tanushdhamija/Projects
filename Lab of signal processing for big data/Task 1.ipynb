# Generating instances of a Random Variable in streaming fashion

Our aim is to implement a function that takes in as input a given distribution (in the form of it's cumulative distribution function (CDF) and it draws random samples that follow the given distribution.

We implement a generator that generates instances of a given random variable $ X $ and its distribution in the form of its Cumulative Distribution Function (CDF) $ F_X $.

This is done by means of ***Inverse Transform Sampling.*** [[1]](https://medium.com/mti-technology/how-to-generate-gaussian-samples-347c391b7959)

We lay down the procedure below :

- Of course samples are not generated from thin air, we start with the uniform random number generator :  $ u ∼ U(0, 1) $ that generates a random number between 0 and 1.
- We then perform the method of inverse transform sampling which converts the uniform random sample $u$ into a random sample of *any* distribution $x$.
- Essentially, we want to find a transformation $ T:[0,1] → R $ such that $ T(u) = x $
- We know that :

$$ F_X(x) = P(X \le x) = P(T(u) \le x) $$

$$ = P(u \le T^{-1}(x)) = T^{-1}(x) $$

$$ \therefore \ x = T(u) = F_X^{-1}(u) $$

- Thus we need to compute the inverse of the CDF and feed it with the uniform random sample to generate a sample which follows the distribution described by $ F_X $
- The inverse of the CDF is carried out in two ways : 
 1. Bisection 
 2. Discretization

```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import islice
from scipy import special
import timeit
import math
```

##  Exponential Distribution


  $$ PDF :  λe^{-λx} $$

  $$ CDF :  1 - e^{-λx} $$

```python
def cdf_exp(x, lmbd=1.0):
    return (1 - np.exp(-lmbd*x))
```

```python
def pdf_exp(x, lmbd=1.0):
    return (lmbd*np.exp(-lmbd*x))
```

```python
plt.figure(figsize=(16,8))
plt.title('Exponential Distribution')
plt.grid(True)
plt.plot(np.linspace(0,10,100), cdf_exp(np.linspace(0,10,100)), label='CDF')
plt.plot(np.linspace(0,10,100), pdf_exp(np.linspace(0,10,100)), label='PDF')
plt.legend()
None
```

##  Standard Normal Distribution


  $$ PDF :  \frac {1}{\sqrt {2 \pi}} e^{-\frac{1}{2} x^2} $$

  $$ CDF :  \frac {1}{2} (1 + erf(\frac {x}{\sqrt{2}})) $$

```python
def cdf_normal(x):
    return 0.5 * (1 + special.erf(x/np.sqrt(2)))
```

```python
def pdf_normal(x):
    return (1.0 / np.sqrt(2*math.pi))*(np.exp(-0.5*(x**2)))
```

```python
plt.figure(figsize=(16,8))
plt.title('Standard Normal Distribution')
plt.grid(True)
plt.plot(np.linspace(-10,10,100), cdf_normal(np.linspace(-10,10,100)), label='CDF')
plt.plot(np.linspace(-4,4,100), pdf_normal(np.linspace(-4,4,100)), label='PDF')
plt.legend()
None
```

### Bisection 

It is one of the simplest root finding algorithm. 
The algorithm applies to any continuous function $ f(x) $ on an interval $ [a,b] $ where the value of the function $ f(x) $ changes sign from $a$ to $b$.
The idea is simple: divide the interval in two, a solution must exist within one subinterval. Select the subinterval where the sign of $ f(x) $ changes and repeat the same procedure again and again until we minimize the interval to a very small point which is an estimate of the root of the function. [[2]](https://www.math.ubc.ca/~pwalls/math-python/roots-optimization/bisection/)

The equation to bisect is going to be : $ F_X(x) - u = 0 $

- Why does the Bisection method work?

In the above equation, which is the function to bisect, $F_X(x)$ is a CDF, so it's range is from 0 to 1. $u$ is a random sample who's range is also from 0 to 1. So we provide a $[a,b]$ in such a way so as to get the minimum (0) and maximum (1) values of the CDF respectively because in that way we are sure that :

$$ F_X(a) - u < 0 $$ 
$$and$$
$$ F_X(b) - u > 0 $$

Hence, a root of the equation exists within the range $ [a,b] $ which is then estimated by the Bisection Method.

By looking at the plots of the CDFs taken as example for demonstration (exponential and standard normal), $ [a,b] $ are taken to be as $ [-100,+100] $

```python
def bisection(func, low, high, maxiter=100):

    assert (func(low) * func(high)) < 0

    for _ in range(maxiter):
        midpoint = (low + high) / 2.0
        if (func(low) * func(midpoint)) > 0:
            low = midpoint
        else:
            high = midpoint

    return midpoint      
```

```python
u = [np.random.uniform(0,1) for _ in range(5000)]

start = timeit.default_timer()
x1 = [bisection(lambda x: cdf_exp(x) - u[i], -100, 100) for i in range(0,len(u))]
plt.figure(figsize=(16,8))
plt.title('Samples generated by Bisection (Exponential)')
plt.plot(np.linspace(0,10,100), pdf_exp(np.linspace(0,10,100)),linewidth=3, label='PDF of exponential')
plt.hist(x1, 100, density=True, label='Samples')
plt.legend()
None

stop = timeit.default_timer()
execution_bisection = stop - start
print(f'Bisection Method took (for exponential CDF){execution_bisection: .2f} seconds')
```

```python
start = timeit.default_timer()

x2 = [bisection(lambda x: cdf_normal(x) - u[i], -100, 100) for i in range(0,len(u))]
plt.figure(figsize=(16,8))
plt.title('Samples generated by Bisection (Standard Normal)')
plt.plot(np.linspace(-4,4,100), pdf_normal(np.linspace(-4,4,100)),linewidth=3, label='PDF of standard normal')
plt.hist(x2, 100, density=True, label='Samples')
plt.legend()
None

stop = timeit.default_timer()
execution_bisection = stop - start
print(f'Bisection Method took (for standard normal CDF){execution_bisection: .2f} seconds')
```

### Discretization

Another method to find the inverse of the CDF is to discretize the CDF. Hence we get two tables, that store the points $x$ and $y$ of the CDF. We can define the number of points we want to use to discretize the CDF with. This method is really fast compared to the Bisection Method (as seen later) but the downside being, if we define a large number of points to discretize which increases the accuracy, but consumes a large amount of memory. So it's an important trade off to keep in mind. Depending on the circumstance and application, we will have to choose one or the other.  

```python
def discretization(func, low, high, sample, num_points=10000):
    
    cdf_x = np.linspace(low, high, num_points)
    cdf_y = func(cdf_x)
    x = cdf_x[np.argmin(abs(cdf_y - sample))]
    
    return x
```

```python
start = timeit.default_timer()
y1 = [discretization(cdf_exp, -100, 100, u[i]) for i in range(0,len(u))]
plt.figure(figsize=(16,8))
plt.title('Samples generated by Discretization (Exponential)')
plt.plot(np.linspace(0,10,100), pdf_exp(np.linspace(0,10,100)),linewidth=3, label='PDF of exponential')
plt.hist(y1, 100, density=True, label='Samples')
plt.legend()
None
stop = timeit.default_timer()
execution = stop - start
print(f'Discretization Method took (for exponential CDF){execution: .2f} seconds')
```

```python
start = timeit.default_timer()
y2 = [discretization(cdf_normal, -100, 100, u[i]) for i in range(0,len(u))]
plt.figure(figsize=(16,8))
plt.title('Samples generated by Discretization (Standard Normal)')
plt.plot(np.linspace(-4,4,100), pdf_normal(np.linspace(-4,4,100)),linewidth=3, label='PDF of standard normal')
plt.hist(y2, 100, density=True, label='Samples')
plt.legend()
None
stop = timeit.default_timer()
execution = stop - start
print(f'Discretization Method took (for standard normal CDF){execution: .2f} seconds')
```

> As can be seen from the execution times of Bisection and Discretization, the discretization method is much faster than the bisection method.

# Mean and Variance Computation

### Welford’s Algorithm : Method to compute mean and variance by considering one sample at a time.

Given a dataset of values, Welford's Algorithm takes one sample at a time and calculates the mean $M$ and variance $V$ of the given dataset.

Since it takes in one sample at a time, we need to setup a stream setting that does the same.

- Inputs : Dataset of samples based on a given distribution, length of the stream
- Outputs : Mean and Variance of the dataset

The algorithm is shown below : [[3]](https://gist.github.com/alexalemi/2151722)

$ input: x = (x_1,...,x_n)^T $

$ m ← x_1, S ← 0 $

$ for \ k ← 2,..., n $

> $ M ← m + \frac {x_k − m}{k} $
>
> $ S ← S +  (x_k − m)(x_k − M) $
>
> $ m ← M $

$ end $

$ V \gets \frac {S}{n−1} $

$ output : M, V $

```python
class MeanVar():
    
    def __init__(self):
        self.mean = 0
        self.counter = 0
        self.sum = 0
        
    def update(self, sample):
        self.counter += 1
        newMean = self.mean + (sample - self.mean) / self.counter
        newSum = self.sum + (sample - self.mean) * (sample - newMean)
        self.mean, self.sum = newMean, newSum
        
        return self.mean, self.sum
```

## Defining the stream

We define a stream which takes in a CDF as input and gives as output a random sample that follows the given distribution. The method of inverse transform sampling is used as was discussed earlier.

```python
def stream(f):
    
    while True:
        cdf_x = np.linspace(-100, 100, 10000)
        cdf_y = f(cdf_x)
        x = cdf_x[np.argmin(abs(cdf_y - np.random.uniform(0,1)))]
    
        yield x
```

### Results and Verification

```python
N = int(input('Enter length of stream: '))
meanvar = MeanVar()
elements=[]

for sample in islice(stream(cdf_normal), N):
    mean, S = meanvar.update(sample)
    elements.append(sample)
    
var = S/(N-1) 

print(f'\nMean (by Welford): {mean: .5f}')
print(f'Variance (by Welford): {var: .5f}')
```

### Comparing the results with the mean and variance obtained through Numpy

```python
mean_np = np.mean(elements)
var_np = np.var(elements)

print(f'Mean (by numpy): {mean_np: .5f}')
print(f'Variance (by numpy): {var_np: .5f}')

mean_error = abs(mean - mean_np) / abs(mean_np)
var_error = abs(var - var_np) / abs(var_np)

print(f'\nError in mean: {mean_error: .1f}%')
print(f'Error in variance: {var_error: .1f}%')
```
