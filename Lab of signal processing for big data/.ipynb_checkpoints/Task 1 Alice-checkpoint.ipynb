#### Given a random variable (RV) and its probability distribution (in the form of cdf cumulative density function), implement a Python generator that generates a instances of the RV in streaming fashion. Each RV instance must be generated by means of the Inverse Transform Sampling.

Random number generators are fundamental for Big Data. Every programming language has a random number generator. They give uniformly distributed in a certain range number. We can scale them to be from $0$ to $1$. If we set a level $ùõº$, the probability that the random number generator is less than $ùõº$ is $ùõº$. 
We compute the probability, then take the random number generator and ask. 
This is the basis of coding: how do we decide which path to take with which probability?

```python
import numpy as np
import math
import matplotlib.pyplot as plt
from itertools import islice
```

Generator

```python
def stream():
    while True:
        yield np.random.uniform(0,1)
```

##### Inverse Transform Sampling 

The Inverse Transform Sampling is the method for generating random sample numbers from any probability distribution given its cumulative distribution function (cdf), an instance of a random variable $x$ with pdf $f_x$, and cdf $F_x$.

Let‚Äôs consider $u\sim U(0,1)$. We want to find a strictly monotone transformation $T:[0,1]‚Ü¶R$ such that $T(u)=x$.

$$F_x(x)=P(x\le x)=P(T(u)\le x)=P(u\le T^{-1}(x))=T^{-1}(x)$$
$$x=T(u)={F_x}^{-1}(u)$$

Then, we need to define the cdf. We choose, for example, the exponential distribution, for which:
- pdf:
$$Œªe^{-Œªx}$$
- cdf:
$$1 - e^{-Œªx}$$

```python
def cdf(x, lmbd=1.0):
    return (1 - np.exp(-lmbd*x))
```

```python
t = np.linspace(0,10,100)
z = [(cdf(i)-0.5) for i in t]
o = [cdf(i) for i in t]


plt.plot(t,z)
plt.plot(t,o)
```

Since $x={F_x}^{-1}(u)$, there is the need to compute the inverse of a cdf. 
We can do it by implementing two methods:
- cdf discretization: discretize the cdf function so that, given a instance $u$, you can approximate the correspondent $x$;
- bisection: given a instance $u$, find the $x$ such that $F_x(x)-u=0$.

##### Bisection Method

```python
def bisection(func, low, high, maxiter=50):

    assert (func(low) * func(high)) < 0

    for _ in range(maxiter):
        midpoint = (low + high) / 2.0
        if (func(low) * func(midpoint)) > 0:
            low = midpoint
        else:
            high = midpoint

    return midpoint  
```

```python
x = []
for sample in islice(stream(), 5000):
    bisect_function = lambda x: cdf(x) - sample
    x.append(bisection(bisect_function,0,100))
    
plt.figure(figsize=(16,8))
plt.title('Samples by Bisection')
plt.hist(x, 100, density=True)
None
```

##### Discretization Method

```python
def discretization(func, low, high, sample, num_points=5000):
    cdf_x = np.linspace(low, high, num_points)
    cdf_y = func(cdf_x)
    x = cdf_x[np.argmin(abs(cdf_y - sample))]
    return x
```

```python
y = []
for sample in islice(stream(), 5000):
    y.append(discretization(cdf, 0, 100, sample))
    
plt.figure(figsize=(16,8))
plt.title('Samples by Discretization')
plt.hist(y, 100, density=True)
None
```

#### Given a stream of instances of a RV, compute mean and variance in a streaming fashion by means of the Welford‚Äôs Algorithm.

Welford‚Äôs Algorithm is a method to compute mean $M$ and variance $V$ by considering one sample $x_i$ at a time.


$$input:\ x=(x_1,...,x_n)^T$$
$$m\gets x_1,\ S\gets0$$
$$for\ k\gets 2,...,n$$
$$do$$
$$M\gets m+\frac{(x_k-m)}{k}$$
$$S\gets S+(x_k-m)(x_k-M)$$
$$m\gets M$$
$$end\ for$$
$$V\gets\frac{S}{n-1}$$
$$output:\ M,V$$

```python
class MeanVar():
    
    def __init__(self):
        self.mean = 0
        self.counter = 0
        self.sum = 0
        
    def update(self, sample):
        self.counter += 1
        newMean = self.mean + (sample - self.mean) / self.counter
        newSum = self.sum + (sample - self.mean) * (sample - newMean)
        self.mean, self.sum = newMean, newSum
        
        return self.mean, self.sum
```

```python
def stream1():
    while True:
        yield np.random.randint(0,100)
```

```python
N = 5000
meanvar = MeanVar()
elements=[]

for sample in islice(stream1(), N):
    mean, S = meanvar.update(sample)
    elements.append(sample)
    
var = S/(N-1) 

print(f'Mean (by Welford): {mean: .5f}')
print(f'Variance (by Welford): {var: .5f}')
```

Comparing the results obtained with the mean and variance obtained through Numpy

```python
mean_np = np.mean(elements)
var_np = np.var(elements)

print(f'Mean (by numpy): {mean_np: .5f}')
print(f'Variance (by numpy): {var_np: .5f}')

mean_error = abs(mean - mean_np) / mean_np
var_error = abs(var - var_np) / var_np

print(f'\nError in mean: {mean_error: .1f}%')
print(f'Error in variance: {var_error: .1f}%')
```
