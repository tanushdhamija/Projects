```python
import numpy as np
import math
import matplotlib.pyplot as plt
from itertools import islice
```

```python
def stream():
    while True:
        yield np.random.uniform(0,1)
```

```python
def samesign(a, b):
        return a * b > 0
```

```python
def cdf(x):
    return (1 - np.exp(-x))
```

```python
x = np.linspace(0,10,100)
e = cdf(x)
f = cdf(x) - 0.5
g = cdf(x) - 0.2
plt.figure(figsize=(12,6))
plt.plot(x,e)
plt.plot(x,f)
plt.plot(x,g)
plt.grid(True)
```

```python
def bisect(func, low, high, maxiter=50):

    assert not samesign(func(low), func(high))

    for i in range(maxiter):
        midpoint = (low + high) / 2.0
        if samesign(func(low), func(midpoint)):
            low = midpoint
        else:
            high = midpoint

    return midpoint      
```

```python
x = []
for sample in islice(stream(), 5000):
    bisect_function = lambda x: cdf(x) - sample
    x.append(bisect(bisect_function,0,100))
    
plt.figure(figsize=(16,8))
plt.hist(x, 100, density=True)
None
```
